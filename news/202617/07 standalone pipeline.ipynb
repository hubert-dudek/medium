Standalone pipelines
Till now, to create a standalone pipeline with one streaming table or materialized view (CREATE OR REFRESH STREAMING and CREATE OR REPLACE MATERIALIZED VIEW) We had to use SQL warehouse, now we can use a general serverless as well.

```sql
CREATE OR REPLACE MATERIALIZED VIEW workspace.default.sales_per_country
CLUSTER BY (country_id)
SCHEDULE EVERY 1 WEEK
AS
SELECT
  country_id,
  SUM(amount) AS total_amount
FROM workspace.default.sales
GROUP BY country_id;
```

```sql
DROP MATERIALIZED VIEW workspace.default.sales_per_country;
```
