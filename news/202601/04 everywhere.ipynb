Literal string coalescing

```sql
CREATE EXTERNAL TABLE main.tmp.demo_ext (raw STRING)
USING DELTA
LOCATION 's3://my-lake/'
         'landing/'
         'app1/'
         '2026/01/04/';
```

Parameter markers

```sql
EXECUTE IMMEDIATE
  'CREATE VIEW main.tmp.v_param AS
   SELECT :val AS c1, :val * 2 AS c2'
USING 10 AS val;
```

```sql
CREATE TABLE t_dec (
  amount DECIMAL(:p, :s)
);
```

```sql
COMMENT ON TABLE my_table IS :comment;
```

```sql
CREATE TABLE t_default (
  c1 INT DEFAULT :def
);
```

Identifiers

```sql
CREATE TABLE t_ident_cols (
  IDENTIFIER(:pk) BIGINT NOT NULL,
  payload STRING
);
```
