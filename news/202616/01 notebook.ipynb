## PySpark Equivalent of "01 My Canvas" Designer Workflow

This notebook replicates the Visual ETL pipeline:
1. **Source** — Read from `hub.default.sale_refund`
2. **Transform** — Filter rows where `customer_id = 101`
3. **Output** — Write results to `hub.default.output_2` (overwrite)

```sql
%python
# Step 1: Read the source table
df_sale_refund = spark.table("hub.`default`.sale_refund")
```

```sql
%python
# Step 2: Apply SQL transformation — filter for customer_id = 101
df_filtered = df_sale_refund.filter("customer_id = 101")
display(df_filtered)
```

```sql
%python
# Step 3: Write to output table (overwrite)
df_filtered.write.mode("overwrite").saveAsTable("hub.`default`.output_2")
```
