```sql
CREATE OR REFRESH STREAMING TABLE bookings_current
CLUSTER BY (booking_date);

CREATE FLOW bookings_current_flow AS
AUTO CDC INTO bookings_current
FROM STREAM workspace.cdf_benchmark_source.bookings WITH (readChangeFeed=true)
KEYS (booking_date, booking_id)
APPLY AS DELETE WHEN _change_type = 'delete'
SEQUENCE BY STRUCT(updated_at, _commit_version)
COLUMNS * EXCEPT (_change_type, _commit_version, _commit_timestamp)
STORED AS SCD TYPE 1;
```
