```python
from pyspark.sql import functions as F

target_table = "source_data"
row_count = 1_000_000

# Optional: increase/decrease depending on cluster size
num_partitions = 2000

base_df = (
    spark.range(0, row_count, 1, num_partitions)
    .withColumn("id", F.expr("uuid()"))
)

# 50 high-cardinality string columns
for i in range(1, 51):
    base_df = base_df.withColumn(
        f"str_{i:03d}",
        F.sha2(
            F.concat_ws(
                "||",
                F.col("id"),
                F.col("id").cast("string"),
                F.lit(i),
                F.rand(seed=i).cast("string"),
                F.monotonically_increasing_id().cast("string")
            ),
            256
        )
    )

# 50 high-cardinality integer columns
for i in range(1, 51):
    base_df = base_df.withColumn(
        f"int_{i:03d}",
        F.abs(
            F.xxhash64(
                F.col("id"),
                F.lit(i),
                F.col("id").cast("string"),
                F.monotonically_increasing_id()
            )
        )
    )

# Drop technical range column
final_df = base_df.drop("id").withColumn("id", F.expr("uuid()"))

# Reorder columns so ID is first
ordered_cols = (
    ["id"]
    + [f"str_{i:03d}" for i in range(1, 51)]
    + [f"int_{i:03d}" for i in range(1, 51)]
)

final_df = final_df.select(*ordered_cols)

# Write into Delta table
(final_df.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(target_table)
)

print(f"Inserted {row_count:,} rows into {target_table}")
```
